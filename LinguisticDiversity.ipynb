Before moving on, lets give a brief overview of the columns in the dataset:

- Country: Name of the country being studied.

- lang_count: The number of languages found in a given country

- lang_count(%): Percentage of the worlds languages that are found in a given country

- lang_count_established/immigrant: These two columns signal how many of the languages in a given country are established, having been in the country for a long time, and immigrant, being recently introduced through immigration.

- nmb_of_speakers_total/mean/median: These three columns gives the total meand and median number of speakers of any language in a given country

- diveristy_index: Contains the diversity index of a given country, by its Greenberg's diversity index, which is the probability of two random speakers in a given country speaking different languages as mother language.

- diversity_coverage: Gives the percentage of the population for which estimates of number of speakers are given. A higher values thus means that the diversity index is calculated based on more complete data

Now, lets move on to looking at linguistic diversity, which just to refresh your memory is defined the likelhood that two random inhabitants of a country have the same mother tongue. 

### Linguistic Diveristy

```python
plt.figure(figsize=(6,6))
ax = sns.boxplot(
    data = df,
    y = 'diversity_index',
    width = 0.3,
    color = 'purple',
    medianprops = dict(color ='green', linewidth=3)
);
plt.ylabel('Diversity Index')
plt.title('Boxplot of linguistic diversity index')
```

The Boxplot for the linguistic diversity index is very neat, showing a median at a little below 0.5 and very evenly sized quartiles. One can notice a very very slight right tail skew however. Lets go ahead and also look at its histogram.

```python

ax = sns.histplot(
    data = df,
    x = 'diversity_index',
    color = 'purple',
)
ax.set_title('Histogram of the linguistic diversity index');
```

We see here, that just as with the number of languages, there are many languages that have very little diversity, but still the distribution is far more even. It is logical to assume that in countries with many languages, the linguistic diversity is high, but this most not necessarily be the fact, as many of the languages could have very few speakers, with just one language being spoken by a clear majority. The discrepancy in the distribution hint towards this being true for the data, but lets plot the two variables against eachother for a better look.


```python
import scipy.stats as stats
fig, ax = plt.subplots(1,1)
sns.regplot(ax = ax, x ='lang_count', y = 'diversity_index', data = df)
ax.set_title( 'Linguistic diversity vs language count')
ax.set_xlabel('number of languages')
ax.set_ylabel('linguistic diversity index')
r, p = stats.pearsonr(df['lang_count'], df['diversity_index'])
ax.text(0.5,0.95,f'r = {r:.3f} p = {p:.4f}', transform=ax.transAxes, ha='left', va='top');

```

Looking at the plot, and considering the calculated r-value, there is indeed a correlation, albeit weak, r = 0.32
Now this plot has included the outliers, so lets do the same thing, but without the outliers that may be skewing the analysis.

```python
filter = df['lang_count'] < 150
filtered_LC = df['lang_count'][filter]
filtered_DI = df['diversity_index'][filter]
fig, ax = plt.subplots()
sns.regplot(ax = ax,x =filtered_LC, y = filtered_DI)
ax.set_title('Linguistic diversity vs language count without outliers')
ax.set_xlabel('number of languages')
ax.set_ylabel('linguistic diversity index')
r, p = stats.pearsonr(df['lang_count'], df['diversity_index'])
ax.text(0.5,0.5,f'r = {r:3f} p = {p:.4f}', transform=ax.transAxes, ha='left', va='top');
```

Looking at the graph without the outliers, the correlation is more clearly visible, but the strength is still weak to moderate with r = 0.32. Looking at the plot, there seems to show characteristics of exponentiality. To make our modell better, lets plot the linguistic diversity against the logarithm of the number of languages.

```python

fig, ax = plt.subplots(1,1)
sns.regplot(ax = ax, x = np.log(df.lang_count), y = 'diversity_index', data = df)
plt.ylabel('linguistic diversity index')
plt.xlabel('the logarithm of the number of languages')
plt.title('linguistic diversity index vs logratihm of the number of languages')
r, p = stats.pearsonr(np.log(df.lang_count), df.diversity_index)
ax.text(0.1,0.95,f'r = {r:3f} p = {p:.4f}', transform=ax.transAxes, ha='left', va='top');
```

Now, this looks even better, and interpreting the correlation estemate, we can conclude that there is a statistically significant and moderate positive relationship between linguistic diversity and the logarithm of the number of languges spoken.


To conclude: A realtionship does indeed exists between the number of languages spoken in a country and its lingusitic diversity, but it doesn't tell the entire tale.

There must be an explanation for this to be found in the data and if we go back to look at the map, that in the western hemisphere (north and south america) there is very low lingusitic diversity, which could have its roots in the colonial history, where spanish, portuguese and english took over, almost exterminating the native americans with the introduction of diseases and forcing away their languages. To really get a better understanding of the influence of geography, we should remove the american continent from the analysis. 

```python
fig, ax = plt.subplots(1,2, figsize = (12,6))

filter = map['continent'].isin(['Europe', 'Africa', 'Asia', 'Oceania'])

sns.regplot(ax = ax[0],
    x = 'latitude', y = 'diversity_index', data = map[filter],
)
ax[0].set_title('Diveristy index vs latitude (south - north) excluding America')
ax[0].set_ylabel('Diversity index')
r1, p1 = stats.pearsonr(map[filter]['latitude'], map[filter]['diversity_index'])
ax[0].text(0.60, 0.95,f'r = {r1:.2f} p = {p1:.3f}', transform=ax[0].transAxes, ha='left', va='top');


sns.regplot( ax = ax[1],
    x = 'longitude', y = 'diversity_index', data = map[map['longitude'] > -50],
)
ax[1].set_title('Diveristy index vs longitde (west - east) excluding America')
ax[1].set_ylabel('Diversity index')
r2, p2 = stats.pearsonr(map[map['longitude'] > -50]['longitude'], map[map['longitude'] > -50]['diversity_index'])
ax[1].text(0.50, 0.95,f'r = {r2:.2f} p = {p2:.3f}', transform=ax[1].transAxes, ha='left', va='top');
fig.suptitle('Linguistic diveristy index and geographical coordinates', fontsize = 15);

plt.tight_layout(pad = 2)
```

With South america removed from the equation, we see that in Eurasia and Africa, there is a significant week to moderate correlation between linguistic index and north/south position, where the further north you go, the less linguisticly diverse the contries become, and we also see that there is no significant correlation between east and west. So there is a relationship between north/south and linguistic diversity, but why? Could this be connected to the climate? To explore this, lets use the distance from the equator as an approximation for climate; the further away, the colder it is.

```python
map['dist'] = np.abs(map.latitude)
fig, ax = plt.subplots(figsize = (12,6))
sns.regplot(ax = ax,
    x = np.abs(map[filter].dist), y = 'diversity_index', data = map[filter],
)
ax.set_title('Diveristy index vs distance to the equator excluding America')
ax.set_ylabel('Diversity index')
ax.set_xlabel('Distance to the equator (absolute latitude)')
r1, p1 = stats.pearsonr(np.abs(map[filter]['latitude']), map[filter]['diversity_index'])
ax.text(0.60, 0.95,f'r = {r1:.2f} p = {p1:.3f}', transform=ax.transAxes, ha='left', va='top');
```

By taking the absolute values, the latitude becomes a measurement of the distance to the equator, and we find a statistically significant moderate negative relationship between climate and linguistic diversity. What does this mean for us? Well it means that the colder it gets, the less linguisticly diverse areas of the world is. Now, having found this correlation, the question is of course, is there a causuality? Has somehow a hotter climate helped languages to establish themselves, without one taking over? While there could be merit to this notion, one should consider the power of society, and the developments of our society which as lagged behind in certain parts of the world. There is a reason that in human geography, it is commonly talked about the global south and the global north, and we concluded that there is a realtionship between linguistic diversity and location on the north-south axis. ### Development factors

The most common and probably most robust predictor of human development is economical development. What we find here is that some factors are more strongly correlated with the others, for example birthrate compared to deathrate. Birtrate correlate more strongly with all other variables, and generally, infant mortality is the strongestly correlated with all other variables. So what is important to take with from this? Well, for our development variables, we have most likely a high degree of mulitcollinearty.

now, lets try to put all of this into relationship with linguistic diversity. Now, we have divided our data into equally large categories with various degree of economic development and linguistic diversity, so a good starting point would be to study their degree of association.

We will start with determining if there is a significant association between them using a chi-square test. The null hypothesis is that there is no association between the variables, and we formulate the alternative hypothesis, that there is an association between economical development and linguistic diversity

```python
tab = pd.crosstab(map.economic_categories, map.diversity_degree)
chi2, p, _, _ = stats.chi2_contingency(tab)
print(chi2, p)
```

As we see, the p value is less than .01, which means that the association is highly significant. We can thus feel confident discarding the null hypothesis, and conclude that there is an association between economical development and linguistic diversity. How large is the effect of this association? To find out, we can calculate cramers V

```python
v = stats.contingency.association(tab, method = 'cramer')
print(f'Cramers V: {v:.3f}')
```

What we find is a fairly small association estimate of 0.25 between economical development and linguistic diveristy, but we do not now the direction of the relationship. Lets see if we go at it from another angel making use of linear regression of the numerical variables GDP per Capita and linguistic diversity index

```python
fig, ax = plt.subplots(figsize=(12,8))

sns.regplot(
    ax = ax,
    data = map,
    x = map.GDP_per_capita,
    y = map.diversity_index,
   
)
r, p = stats.pearsonr(map.GDP_per_capita,map.diversity_index)
print(r, p)
```

The regression modell intuitivly confirms what we concluded using cramers V, the association between economical development and linguistic developments is weak to moderate, but what we also see is that the direction of the realtionships is negative, which means that as countries get richer, they tend to get less linguistically diverse. 

Now, we have lots of other development factors aswell, so what we will do is to plot them all against linguistic diversity and look at the relationships.

```python
for factor in develop_var:
    fig, ax = plt.subplots(figsize = (5,5))
    sns.regplot(
            x = factor,
            y = 'diversity_index',
            data = map,
        )
    ax.set_title(f'Diveristy index vs {factor}')
    ax.set_ylabel('Diversity index')
    r1, p1 = stats.pearsonr(map[factor], map['diversity_index'])
    ax.text(0.60, 0.95,f'r = {r1:.2f} p = {p1:.3f}', transform=ax.transAxes, ha='left', va='top');
```

Now, looking at these graps and considering the calculated correlation estimates, we see that the correlation varies across the variables, but always significant and always weak to moderate. As we can see, the strongest correlation is with literacy, infant mortality and birthrate, approximation r = 0.5. Interestingly, the weakest correlation is with GDP_per_capita. So what can we conclude from this. Well, we can conclude, that there is indeed a correlation between development factors linguistic diveristy, meaning that as countries get more developed, this happens at the expense of linguistic diversity.

As an important assumption for a linear modell is that there is no multicollinearity, we will calculate the variance inflation factor for all predictor variables, and if it is larger than 10, we will remove it. 

```python

from statsmodels.stats.outliers_influence import variance_inflation_factor 
vif_data = pd.DataFrame() 
vif_data["feature"] = map[predictors].columns 
  
# calculating VIF for each feature 
vif_data["VIF"] = [variance_inflation_factor(map[predictors].values, i) 
                          for i in range(len(map[predictors].columns))] 
  
print(vif_data)
```
